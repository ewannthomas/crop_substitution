```python
import pandas as pd
from pathlib import Path
```

```python
dir_path = Path.cwd().parents[2]

data_folder = dir_path.joinpath("data/ibis/raw_manually_sorted")

df_path = data_folder.joinpath("Captive_and_Merchant_Power_Plant.xlsx")

export_path = dir_path.joinpath("data/ibis/interim/Captive_and_Merchant_Power_Plant.csv")
```

```python
df = pd.read_excel(df_path)
df
```

```python
df.columns
```

### Filtering Plants

```python
df['AddCode'].unique()
```

```python
df = df[df['AddCode'].isin(['Plant','Hydel Power Plant'])]
# df.drop('AddCode',axis=1, inplace=True)
```

### Check Duplicates

```python
df[df.duplicated(keep=False)]
```

```python
isid_cols = ['Company', 'Pincode', 'Add1', 'Add2']


df[df.duplicated(isid_cols,keep=False)]
```

### Cleaning Column names and Columns

```python
cols = df.columns

cols = [x.replace(" ", "_").lower() for x in cols]

cols
```

```python
df.columns = cols

df.rename(columns={
    "power_plant_capacity":'capacity',
    'city':'units',
    "upcoming_power_plants":'upcoming_capacity',
    "taluk":"upcoming_capacity_units",
    "citypost":"city",
    "taluk2":'taluk',
    "units.1":"upcom_capa_unit"
}, inplace=True)
```

```python
df.drop(['oldname', 'groupname', "addcode", "tel", "fax", "email-1", "email-2", "email-3"
], axis=1, inplace=True)
```

```python
df.to_csv(export_path, index=False)
```
