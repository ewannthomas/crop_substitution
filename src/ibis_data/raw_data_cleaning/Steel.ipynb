```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
dir_path = Path.cwd().parents[2]

data_folder = dir_path.joinpath("data/ibis/raw_manually_sorted")

df_path = data_folder.joinpath("Steel_Capacity_by_Furnaces.xlsx")

export_path = dir_path.joinpath("data/ibis/interim/Steel.csv")
```

```python
df = pd.read_excel(df_path)
df
```

```python
df.columns
```

### Filtering Plants

```python
df['Addcode'].unique()
```

```python
df = df[df['Addcode'].isin(['Plant'])]
df.drop('Addcode',axis=1, inplace=True)
```

### Check Duplicates

```python
df[df.duplicated(keep=False)]
```

```python
isid_cols = ['Company', 'Pincode', 'Add1', 'Add2']


df[df.duplicated(isid_cols,keep=False)]
```

### Cleaning Column names and Columns

```python
cols = df.columns

cols = [x.replace(" ", "_").lower() for x in cols]

cols
```

```python
df.columns = cols

df.rename(columns={
    
   "units":"unit",
   "unit":"upcoming_capacity_unit",
   "upcpy":'upcoming_capacity',
   'upcpyunits':'upcoming_capacity_units',
   'citypost':'city'
   
}, inplace=True)
```
