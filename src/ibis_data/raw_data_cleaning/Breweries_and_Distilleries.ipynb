```python
import pandas as pd
from pathlib import Path
```

```python
dir_path = Path.cwd().parents[2]

data_folder = dir_path.joinpath("data/ibis/raw_manually_sorted")

df_path = data_folder.joinpath("Breweries_and_Distilleries.xlsx")

export_path = dir_path.joinpath("data/ibis/interim/Breweries_and_Distilleries.csv")
```

```python
df = pd.read_excel(df_path)
df
```

```python
df.columns
```

### Filtering Plants

```python
df['AddCode'].unique()
```

```python
df = df[df['AddCode'].isin(['Plant', 'unit', 'plant'])]
# df.drop('AddCode',axis=1, inplace=True)
df
```

### Check Duplicates

```python
df[df.duplicated(keep=False)]
```

### Cleaning Column names and Columns

```python
cols = df.columns

cols = [x.replace(" ", "_").lower() for x in cols]

cols
```

```python
df.columns = cols

df.rename(columns={
    "units":"upcoming_capacity_unit"
}, inplace=True)
```

```python
df.drop(['tel',
 'fax',
 'email',
 'names',
 'captive_power_plant_capacity'], axis=1, inplace=True) 
```

```python
df
```

```python
df.to_csv(export_path, index=False)
```
