```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
dir_path = Path.cwd().parents[2]

data_folder = dir_path.joinpath("data/ibis/compiled")

df_path = data_folder.joinpath("power_plants.csv")

power_dups_path = dir_path.joinpath("data/ibis/data_docs/power_plant_compiled_duplicates.csv")

export_path = dir_path.joinpath("data/ibis/processed/power_plants.csv")
```

```python
df = pd.read_csv(df_path)

df.dtypes
```

```python
index_cols = ['company_name','add1', 'add2',
       'city', 'pincode', 'taluk', 'district', 'state' ]


df[df.duplicated(index_cols, keep=False)].sort_values(index_cols)

# .to_csv("temp.csv", index=False)
```

## Removing Inter Directory Duplicates

Call in the `dups` file where duplicates are marked by 0 in the dups column

```python

dups_cols = index_cols + ['dups', 'directory']

dups_df = pd.read_csv(power_dups_path)[dups_cols]

dups_df = dups_df[~dups_df['dups'].isna()]

dups_df
```

```python
dups_df[dups_df.duplicated(keep=False)]
```

```python
dups_df.drop_duplicates(keep='first', inplace=True)
```

```python
merge_cols = index_cols + ['directory']
```

```python
df_new = df.merge(dups_df, on=merge_cols, how='outer', validate="m:1")
df_new = df_new[df_new['dups'].isna()].drop('dups', axis=1)

df_new
```

```python
df_new.to_csv(export_path, index=False)
```
